Python String

Creation of String

```python
# string can be craeted using single & double quotes

s1 = 'Shilpa'
s2 = "Shilpa"

print(s1)
print(s2)

# Multi line string can be created using triple quotes

triple_quote_str = '''
This is a multi line string
created using triple quotes.
'''

print(triple_quote_str)
```

```python
# Accessing characters in a string using indexing

''' 
Strings are indexed sequences, 
which means positive index starts from the left, 
negative indices starts from the right side. 
'''

name = "ShilpashreeRamanjinappaRaj"

print(name[0])
print(name[-1])
print(name[1:7:2])
```

```python
print(name[::-1])
```

Slicing

```python
s = "Shilpashree Raj"

print(s[6:11])
```

```python

for char in name:
    print(char, end=',')
```

```python
s = 'shilpa'
print(s)
s = 'S' + s[1:]
print(s)
```

```python
# Deleting a string variable
del s
```

```python
s = 'shilpa Raj'
print(s)
s1 = 'S' + s[1:]
print(s1)
s2 = s1.replace('Shilpa', 'Shilpashree')
print(s2)

del s
del s1
del s2
```

Common String methods

1. len() function to get the length of a string
2. str() function to convert other data types to string
3. upper() and lower() methods to change the case of a string
4. split() method to split a string into a list of substrings
5. join() method to join a list of strings into a single string
6. find() method to find the index of a substring in a string
7. replace() method to replace a substring with another substring

```python
str1 = "Hello"

print(len(str1))
```

```python
valuee = 123
print(type(valuee))

valuee = str(valuee)
print(type(valuee))
```

```python
print(str1.upper())
print(str1.lower())
```

```python
name = "  ShilpashreeRamanjinappaRaj  "
print(name)
print(name.strip())

s = 'python is fun'
print(s)
print(s.replace('fun', 'awesome'))
```

Concatenation and Repeated Strings

```python
s1 = "Hello "
s2 = "World..!"

s3 = s1 + s2
print(s3)
print(s1 + s2)
print(s1 * 3, end=' ')
```

Formatting Strings

```python
# 1. Using f-strings

name = 'shilpa'
age = 30
city = 'Bangalore'

print(f"My name is {name}, I am {age} years old and I live in {city}.   ")

# 2. Using the format() method

print("My name is {}, I'am {} years old and I live in {}.".format(name, age, city))

```
