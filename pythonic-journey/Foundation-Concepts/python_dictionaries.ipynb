Dictionary
1. Stored information in KEY-VALUE pairs
2. Keys must be always unique and immutable
3. Whcih makes dicts - ideal for accessing data by a specific name rather than numeric position

```python
data = { "name": "Jake", "age": 22 }
print(data)
```

Creating a dictionary

a dict is created by  writing key-value pairs inside { }, where each key is connected to a value using colon (:)

```python
dict1 = {1:'one', 2:'two', 3:'three'}
print(dict1)
```

```python
dict2 = dict(a = "one", b = "two", c = "three")
print(dict2)
```

Accessing items in dictionary

```python
print(dict2["a"])

print(dict2.get("b"))
```

Adding and Updating dict items

```python
dict1['d'] = "four"

print(dict1)
```

```python
dict1['1'] = "ONE"

print(dict1)
```

Iterating through a dictionary

```python
d = {1:'one', 2:'two', 3:'three', 4:'four', 5:'five', 6:'six', 7:'seven', 8:'eight', 9:'nine', 10:'ten'}
for key in d:
    print(key, end = " ")
print()

for value in d.values():
    print(value, end = " ")
print()

for key, value in d.items():
    print(key, "-->", value, end = ", ")
print()
```

Nested Dictionaries

```python
d = {1:'one', 2:'two', 3:'three', 4 : {"one":1, "two":2, "three":3, "four":4, "five":5}, 5 : 'five'}

print(d)
print(d[4])
```
