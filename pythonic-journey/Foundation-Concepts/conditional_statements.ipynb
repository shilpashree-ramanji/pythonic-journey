CONDITIONAL STATEMENTS IN PYTHON

```python
# IF conditinal statement

x = 10

if x > 5:
    print("X is greater than 5")
```

```python
# Short Hand IF

age = 20
if age >=18: print("Eligible for voting..!!")
```

```python
# IF ELSE Conditional Statement

age = 17

if age >=18:
    print("ELigible to vote")
else:
    print("Not Eligible to vote")

```

```python
# Short Hand IF ELSE

age = 10
result = "Ticket is free" if age <= 5 else "Pay the ticket"
print(result)
```

```python
# elif Statement

age = 1

if age <=3:
    print("Baby")
elif age >3 and age <=12:
    print("child")
else:
    print("Adult")

print("End of the program")
```

```python
# Nested IF - ELSE Statement

age = 2
is_member = False

if age >=18:
    if is_member:
        print("Eligible for 50% discount")
    else:
        print("ELigible for 10% discount")
else:
    print("Not eligible for discount")

```

```python
# Ternary conditional statement

age = 20

result = "Yound Adult" if age <30 else "adult"
print(result)
```

```python
# Match Case Statement

Number  = 1

match Number:
    case 1:
        print("One")
    case 2 | 3:
        print("Two or Three")
    case 3:
        print("Three")
```

LOOPS IN PYTHON

```python
# FOR LOOP

n = 4
for i in range(n):
    print(i, end = " ")
```

```python
# Iteration over List, Tuple, String, and dictionary using for loops

List1 = ["apple", "banana", "cherry"]
for i in List1:
    print(i, end = " ")
print()

Tuple1 = ("Appale","banana", "cherry")
for i in Tuple1:
    print(i, end = " ")
print()

String1 = "Shilpashreeraj"
for i in String1:
    print(i, end = " ")
print()

Dict1 = {"name": "shilpa", "age": 20, "city": "Bangalore"}
for key in Dict1:
    print(key, ":", Dict1[key], end = " ")
print()
for key, value in Dict1.items():
    print(key, ":", value, end = " ")
print()

set1 = {"apple", "banana", "cherry"}
for i in set1:
    print(i, end = " ") 
print()
```

```python
# Iterating by index using for loop
List1 = ["apple", "banana", "cherry"]
for i in range(0, len(List1)):
    print(List1[i], end = "\t")
    print(i, end = "\t")
```

NESTED LOOPS

LOOP CONTROL STATEMENTS

```python
# Continue statement
# The continue statement is used to skip the current iteration of a loop and move to the next iteration. 
# It is often used when you want to skip certain conditions or values in a loop.

for letter in "geeksforgeeks":
    if letter == "e" or letter == "s":
        continue
    print("Current Letter:", letter, end = "\n")
```

```python
# break statement
# The break statement is used to exit a loop prematurely when a certain condition is met.
for letter in "geeksforgeeks":
    if letter == 'k':
        break
    print("Current Letter:", letter, end = "\n")
```

```python
# Pass Statement
#The pass statement is a null statement that is used as a placeholder when a statement is required syntactically but you do not want to execute any code. 
# It is often used in situations where we want to define a function or a class but do not want to implement it yet, or when we want to create an empty loop or conditional statement.

for letter in "geeksforgeeks":
    pass
print("Current Letter:", letter, end = "\n")

#It will print the last letter of the string "geeksforgeeks" which is "s" because the loop iterates through each letter in the string and assigns it to the variable "letter". After the loop completes, the variable "letter" will hold the value of the last letter in the string, which is "s".
```

```python
for num in range(10, 14):
    for i in (range(2, num)):
        if num % i == 1:
            print(num)
            break
```
