```python
# Typecode during initialization to define C-type

import array as arr

arr1 = arr.array('i', [1, 2, 3, 4, 5])
print(arr1)
print(type(arr1))
```

```python
for i in range(0, len(arr1)):
    print(i, end = ',')
    print(arr1[i], end = '\n')
```
