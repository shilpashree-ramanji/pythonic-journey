TUPLES

Immutable ordered collection of elements.
Tuples are similar to list but immutable
tuples can hold diff types data types


Creating a tuple --> can be created by placing all the items inside paranthesis(), seperated by comma.

```python
# Empty tuple

tup1 = ()
print(tup1)

tup2 = ('shilpa', 'shree', 'raj')
print(tup2)

tup3 = tuple('shilpa')
print(tup3)

list1 = ['shilpa', 'shree', 'raj']
tup4 = tuple(list1)
print(tup4)
```

```python
# accessing tuple elements

tup = tuple("Shilpashree Raj")

print(tup[1::2])
```

```python
# Tuple unpacking

tup = ('shilpa', 'shree', 'raj')

a, b, c = tup
print(a)
print(b)
print(c)
```

```python
# Concatenation of tuples

tup1 = (1, 2, 3)
tup2 = ('shilpa', 'shree', 'raj')

tup3 = tup1 + tup2
print(tup3)
```

```python
# Slcicing of tuples

tup = (1, 2, 3, 4, 5, 6)

print(tup[1:5])
```

```python
# deleting a tuple
tup = (1, 2, 3, 4, 5, 6)
print(tup)
del tup
print(tup)  # This will raise an error since the tuple has been deleted
```

```python
# Tuple unpacking with * operator

tup = (1, 2, 3, 4, 5)
a, *b, c = tup
print(a)  # Output: 1
print(b)  # Output: [2, 3, 4]
print(c)  # Output: 5
```
